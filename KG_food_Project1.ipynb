```python
import re

def convert_to_float(fraction_string):
    fraction_match = re.match(r'(\d+)\s+(\d+)/(\d+)', fraction_string)
    if fraction_match:
        whole_number = int(fraction_match.group(1))
        numerator = int(fraction_match.group(2))
        denominator = int(fraction_match.group(3))
        fraction_value = whole_number + (numerator / denominator)
        return fraction_value
    else:
        return None

```
